```python
from pici import Pici
import pandas as pd
from pici.preprocessors.posts import preprocessed_text
import networkx as nx
```

```python
pici = Pici(
    cache_dir='../../../cache',
    start='2017-01-01',
    end='2019-01-01',
    cache_nrows=3000
)
```

```python
osm = pici.communities['OpenStreetMap']
community = osm
```

```python
level, df = preprocessed_text(osm)
```

```python
df['date'] = community.posts[community.date_column]
text_col='preprocessed_text__words_no_stop'
df['t'] = df[text_col].apply(lambda l: tuple(str(t).lower() for t in l))
```

```python
df
```

```python
from textacy.representations.network import build_similarity_network

g = build_similarity_network(df['t'].tolist(), edge_weighting='token_sort_ratio')
```

```python
g.nodes
```

```python
filtered_df = df[df['date'] < '2018-11-15 '].drop_duplicates(subset='t')
filtered_texts = set(filtered_df['t'])
filter = lambda n: n in filtered_texts
sub_g = nx.subgraph_view(g, filter_node=filter)
```

```python
adj = nx.adjacency_matrix(g, weight='weight')
name_to_similarities = {k:adj[i] for k,i in zip(sub_g.nodes, range(0,sub_g.number_of_nodes()))}
```
